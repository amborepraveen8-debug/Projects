### Exercise 1: Prime Numbers

```python
num = int(input("Enter number to check prime or not"))
if num <= 1:
    print(f"{num} is not prime number")
else:
    count = 0
    for i in range(1, num+1):
        if num%i == 0:
            count = count+1
    if count == 2:
        print("prime Number")
    else:
        print("Not prime number")
```

### Exercise 3: Squares of Even/Odd Numbers

```python
userin = input("chose even or odd")
userin = userin.strip().lower()
if userin == "even":
    for i in range(100, 200):
        if i%2 == 0:
            print(f"{i}: {i*i}")
elif userin == "odd":
    for i in range(100, 200):
        if i%2 != 0:
            print(f"{i}: {i*i}")
else:
    print("Enter valid input")
        
```

### Exercise 4: Hospital Billing System

```python
rc_day = 2000
dr_fee = 1500
lab_ch = 300
med_ch = 2400
num_admit = 4
num_labt = 3
room_total = rc_day * num_admit
lab_total = lab_ch * num_labt
subtotal = room_total + dr_fee + lab_total + med_ch
if subtotal > 10000:
    discount = subtotal * 0.10
else:
    discount = 0

final_bill = subtotal - discount
print("Hospital Bill")
print(f"Room Charges       : ₹{room_total}")
print(f"Doctor Fee         : ₹{dr_fee}")
print(f"Lab Tests Charges  : ₹{lab_total}")
print(f"Medicine Charges   : ₹{med_ch}")
print("---------------------")
print(f"Subtotal           : ₹{subtotal}")
print(f"Discount           : ₹{discount}")
print("========")
print(f"Final Bill Amount  : ₹{final_bill}")
```
