### Hospital Patient Data Analysis

```python
# load the pandas library
import pandas as pd
```

```python
# load the dat set 
patient_data = pd.read_csv('/DataScienceNotes/Assignments/04-Basic Stats-2/Patient_Data.csv')
```

```python
# check the loaded data
patient_data.head()
```

```python
# get information of data set find any null values in data set
patient_data.info()
# found 2 null values in the BillAmount column
```

```python
# fill the null values with mean of BillAmount column
BA_mean=patient_data['BillAmount'].mean()
patient_data['BillAmount']= patient_data['BillAmount'].fillna(BA_mean)
patient_data
```

```python
# # Identifying the numerical column and,
# check the min, max, standerd diviation, mean and quaterly values 
patient_data.describe()
```

```python
# Drop the columns not required for the analysis
pt_data= patient_data.drop(['ReceptionistID', 'CheckInTime'], axis=1)
pt_data
# axis=1 refers to column, 0 refers to row
```

```python
# remove duplicates from the PatienID column
pt_data.drop_duplicates(subset=['PatientID'], keep='first', inplace=True)
pt_data
```

```python
# load the billing data set
billing_data = pd.read_csv('/DataScienceNotes/Assignments/04-Basic Stats-2/Billing_Data.csv')
billing_data.head()
```

```python
# merge patient data and billing data using patient id
final_patient_biiling = pd.merge(pt_data, billing_data, on='PatientID', how = 'outer' )
final_patient_biiling
# this will also concat both patient data and billing data.
# merged both data set as outer to get all the data from both the data set.
```

```python
# create new patient data 
new_patient_data = [['114', 'Aix', 'cardiology', 'Dr.Lee', 3000, 1000, 2000],['115', 'bix', 'cardiology', 'Dr.Rose', 5000, 1000, 4000]]
# creating the data frame of the  new patient data
df_new_pt = pd.DataFrame(new_patient_data, columns = ['PatientID', 'Name', 'Department', 'Doctor', 'BillAmount', 'InsuranceCovered', 'FinalAmount'])
# add new patient data to the data set
final_billing=pd.concat([final_patient_biiling,df_new_pt], ignore_index=True)
final_billing
```
